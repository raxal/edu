__1.__ Даны прямая $l$ и точка $P$. С какой стороны $P$ лежит от прямой?

Пусть даны $l \colon Ax + By + C = 0$ и $P(x',y')$.

Вопрос неполный и нам нужен какой-то ориентир.
Так вот, у прямой есть нормальный вектор $\vec n (A,B)$.
Отложу $\vec n$ от точки на прямой и буду определять лежит ли $P$ по одну сторону с этим концом или нет.

Пусть $M(x_0, y_0) \in l$.

*Заметим, что тогда $A x_0 + B y_0 + C = 0$ и свободный коэффициент равен $C = -A x_0 - B y_0$.*

Тогда если угол между $\vec{MP}(x'-x_0, y'-y_0)$ и $\vec n$ острый, то $P$ и конец $\vec n$ в одной полуплоскости; если угол тупой --- в разных.

Считаем, $\vec{n} \cdot \vec{MP} = A (x' - x_0) + B (y' - y_0) = A x' + B y' - A x_0 - B y_0 = A x' + B y' + C$ и по знаку определяем знак $\cos \phi$.

*В общем случае*, получаем короткое правило:

Если $l \colon Ax + By + C = 0$ и $P(x',y')$, то 
1. $Ax' + By' + C > 0$ $\Rightarrow$ точка $P$ в полуплоскости, в которую указывает $\vec n$
2. $Ax' + By' + C < 0$ $\Rightarrow$ точка $P$ и конец $\vec n$ в разных полуплоскостях
