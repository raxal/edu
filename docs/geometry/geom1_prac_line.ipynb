__1.__ Допустим нужно составить уравнение прямой $l$.
Для этого нам понядобится точка на этой прямой и вектор, который этой прямой параллелен.

Пусть нужно составить уравнение прямой $AB$, где $A(1,3)$, $B(2,-1)$.

В качестве точки возьмем $A$ и в качестве вектора $\vec{AB} (1,-4)$.

Тогда параметрическое уравнение $l$:
$\left\{\begin{aligned}
x &= t + 1\\
y &= -4t +3
\end{aligned}\right.$

__2.__ Пусть теперь дано параметрическое уравнение и мы хотим общее уравнение $l$.
Нам нужно решить одно из уравнений относительно параметра $t$ и подставить его в другое.

Из нашего уравнения выражаем: $t = x - 1$ и подставляем $y = -4(x-1) + 3$.

Преобразовав получаем общее уравнение: $4x + y -7 = 0$.

__3.__ Теперь как из общего уравнения $l$ получить параметрическое.
Нужно снова получить точку прямой и направляющий вектор.

Точку получаем взяв любое решение общего уравнения: при $x = 0$ получим $y = 7$.
Точка $P (0, 7) \in l$.

А в качестве вектора из уравнения вида $Ax + By + C = 0$ можно взять $\vec a(-B,A)$.
То есть у нас $\vec a(-1, 4)$.

Все, получаем параметрическое уравнение $l$:
$\left\{\begin{aligned}
x &= -t\\
y &= 4t + 7
\end{aligned}\right.$

__4.__ Пусть нужно опустить из некой точки $P (2,6)$ на $l: 4x + y -7 = 0$ перпендикуляр.

Нам известно, что коэффициенты общего уравнения - координаты перпендикулярного прямой вектора.
Поэтому составляем уравнение прямой, проходящей через $P$ параллельно $\vec a (4,1)$:
$\left\{\begin{aligned}
x &= 4t + 2\\
y &= t + 6
\end{aligned}\right.$
