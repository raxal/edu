Для составления уравнения плоскости или прямой нужна _точка_ и нужны _направляющие векторы_:
для плоскости два ветора, для прямой --- один (как и ранее).

__1.__ Составить уравнение плоскости $ABC$, если известны координаты $A(0,1,2)$, $B(3,4,5)$ и $C(4,3,1)$.

Это плоскость, проходящая через точку $A(0,1,2)$ параллельно векторам $\vec{AB}$, $\vec{AC}$:  
находим $\vec{AB}(3, 3, 3)$ и $\vec{AC}(4, 2, -1)$, и записываем _параметрическое уравнение_
$$\left\{
\begin{aligned}
x &= 3t_1 + 4t_2\\
y &= 3t_1 + 2t_2 + 1\\
z &= 3t_1 - t_2 +2
\end{aligned}
\right.$$

Если хотим общее уравнение, то решим любые два из них относительно $t_1, t_2$ и подставим в третье оставшееся уравнение:  
например, из первого $t_1 = \frac13 x -  \frac43 t_2$,  
подставим во второе $y = 3(\frac13 x -  \frac43 t_2) + 2t_2 + 1$,  
отсюда $t_2 = \frac12 x - \frac12 y + \frac12$,
подставим $t_2$ обратно, $t_1 = \frac13 x -  \frac43 (\frac12 x - \frac12 y + \frac12) = -\frac13 x + \frac23 y - \frac23$;
теперь подставляем все в третье, $z = 3(-\frac13 x + \frac23 y - \frac23) - (\frac12 x - \frac12 y + \frac12) + 2$, ..., $\frac32 x - \frac52 y + z + \frac12 = 0$, умножим на 2 и запишем _общее уравнение_
$$3 x - 5 y + 2z + 1 = 0$$

__2.__ Написать параметрическое уравнение плоскости $3 x - 5 y + 2z + 1 = 0$.

Знать параметрическое уравнение все равно, что знать точку и направляющие векторы, как мы видим.  
Поэтому можно взять точку --- любое решение уравнения и пару решений соответствующего однородного уравнения в качестве направляющих векторов.
Но векторы нужно брать линейно независимые (не параллельные друг другу, иначе они не будут однозначно определять плоскость).

А можно не переживать обо всем этом и поступить проще:  
пусть $x = t_1$, а $y = t_2$, тогда $z = -\frac 32 t_1 + \frac 52 t_2 - \frac 12$ или
$$\left\{
\begin{aligned}
x &= t_1\\
y &= t_2\\
z &= -\frac 32 t_1 + \frac 52 t_2 - \frac 12
\end{aligned}
\right.$$

__3.__ Написать направляющий вектор прямой, заданной общим уравнением
$$\left\{
\begin{aligned}
2x + y - z &= 0\\
-y + 2z &= 0
\end{aligned}
\right.$$
Например, чтобы определить взаимное расположение этой прямой с какой-то другой.

Это уравнение --- уравнение прямой как пересечение двух плоскостей.  
Нормальные векторы этих плоскостей: $\vec n_1 (2,1,-1)$ и $\vec n_2 (0,-1,2)$.
Вектор $\vec n_1$ перпендикулярен любому вектору, параллельному первой плоскости, в том числе и направляющему вектору нашей прямой. Вектор $\vec n_2$ перпендикулярен любому вектору, параллельному второй плоскости, в том числе направляющему вектору прямой пересечения.
Нам нуже ветор, перпендикулярный $\vec n_1$ и $\vec n_2$.
Поэтому возьмем $\vec v = [\vec n_1, \vec n_2]$.

Считаем и получаем вектор, параллельный данной прямой: $\vec v (1, -4, -2)$.
